```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from text files
data1 = np.loadtxt('/content/Data_1.txt')
data2 = np.loadtxt('/content/Data_2.txt')

# Function to find maxima and minima
def find_peaks(signal):
    maxima = []
    minima = []

    for i in range(1, len(signal) - 1):
        if signal[i] > signal[i - 1] and signal[i] > signal[i + 1]:
            maxima.append(i)
        if signal[i] < signal[i - 1] and signal[i] < signal[i + 1]:
            minima.append(i)

    return maxima, minima

# Find peaks for both datasets
maxima1, minima1 = find_peaks(data1)
maxima2, minima2 = find_peaks(data2)

# Plotting the data
def plot_signal(signal, maxima, minima, title):
    plt.figure(figsize=(10, 6))
    plt.plot(signal, label='Signal')
    plt.scatter(maxima, signal[maxima], color='red', label='Maxima')
    plt.scatter(minima, signal[minima], color='blue', label='Minima')
    plt.title(title)
    plt.xlabel('Index')
    plt.ylabel('Signal Value')
    plt.legend()
    plt.show()

# Plot the results
plot_signal(data1, maxima1, minima1, 'Data 1 Signal with Peaks')
plot_signal(data2, maxima2, minima2, 'Data 2 Signal with Peaks')

# Print indices of maxima and minima
print("Data 1 Maxima Indices:", maxima1)
print("Data 1 Minima Indices:", minima1)
print("Data 2 Maxima Indices:", maxima2)
print("Data 2 Minima Indices:", minima2)
```
